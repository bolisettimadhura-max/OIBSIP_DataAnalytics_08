```python
import pandas as pd

data = {
    'App': ['WhatsApp', 'Instagram', 'Facebook', 'Spotify', 'YouTube', 'Telegram', 'Snapchat', 'Twitter', 'Zoom', 'Google Maps'],
    'Category': ['Social', 'Social', 'Social', 'Music', 'Video', 'Communication', 'Social', 'Social', 'Business', 'Navigation'],
    'Rating': [4.5, 4.4, 4.2, 4.6, 4.7, 4.5, 4.1, 4.0, 4.6, 4.4],
    'Reviews': [1000000, 900000, 850000, 700000, 1200000, 600000, 500000, 450000, 300000, 400000],
    'Installs': [5000000, 4500000, 4000000, 3500000, 6000000, 3000000, 2500000, 2000000, 1500000, 1800000]
}

df = pd.DataFrame(data)

print("Dataset Loaded Successfully")
df.head()
```

```python
# Dataset Information

print("Dataset Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

```python
# Statistical Summary

df.describe()
```

```python
# Missing Values

print(df.isnull().sum())
```

```python
# App Ratings

plt.figure(figsize=(8,5))

sns.barplot(x='App', y='Rating', data=df)

plt.title("App Ratings Comparison")
plt.xticks(rotation=45)

plt.show()
```

```python
# App Installs Analysis

plt.figure(figsize=(8,5))

sns.barplot(x='App', y='Installs', data=df)

plt.title("App Installs Comparison")
plt.xticks(rotation=45)

plt.show()
```

```python
# Category Distribution

df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("App Category Distribution")
plt.ylabel("")

plt.show()
```
